handle missing values

fill missing numerical values with mean(average)

```python
df['Years of Experience'] = df['Years of Experience'].fillna(df['Years of Experience'].mean())
df
```

fill missing categorical values

```python
df['Location'] = df['Location'].fillna('Unknown')
df   #in this table we donot have any missing categorical valiues
```

remove duplicate rows

```python
df = df.drop_duplicates()
df    # if number of row is same as shet then there is no duplicate row in sheet
```

check and fix data types

```python
df.dtypes

df['Age Group'] = df['Age Group'].astype(int)
df
```

rename colume options

```python
df = df.rename(columns={
    "Years of Experience":"YOF",
    "Location":"LOC"



})
df
```

handle outliers( simple filtering )

```python
df_sort = df[df['Monthly Salary (INR)'] < 10000]
df_sort.head()
{"repr_error":"0","type":"dataframe","variable_name":"df_sort"}
```

clean text data

```python
#remove extra space from the beginning and of each Age.
df['Education'] = df['Education'].str.strip()

#convert all  names to lower case letters.
df['LOC'] = df['LOC'].str.lower()
```
